```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv("https://raw.githubusercontent.com/asifraza06/Excercise/master/codebasics%20Hindi/2%20-%20Linear%20Regression%20Exercise/canada_per_capita_income.csv")
```

```python
df.head()
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per_capita_income')
plt.scatter(df.year,df.per_capita_income,color='red',marker='+')
```

```python
new_df = df.drop('per_capita_income',axis = 'columns')
new_df
```

```python
model = linear_model.LinearRegression()
model.fit(new_df,df.per_capita_income)
```

```python
model.predict([[2020]])
```
